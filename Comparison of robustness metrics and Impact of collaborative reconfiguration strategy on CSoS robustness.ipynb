```python
x = np.linspace(0, 1, len(PMA))
fig, ax1 = plt.subplots()

# 绘制 PMA 数据，使用左侧纵轴
color = 'tab:blue'
# ax1.set_xlabel('Attack nodes proportion')
# ax1.set_ylabel('Normalized Operation Capability', color=color)
ax1.plot(x, PMA, linestyle='-', marker='o', markersize=3, label='PMA_Operation Capability', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylabel('Operation Capability', color=color)  # 设置左侧纵坐标标签

# 创建第二个纵轴，共享 x 轴
ax2 = ax1.twinx()  

# 绘制 PMA_C 数据，使用右侧纵轴
color = 'tab:red'
# ax2.set_ylabel('Normalized Connectivity', color=color)
ax2.plot(x, PMA_C, linestyle='-', marker='d', markersize=3, label='PMA_Connectivity', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel('Connectivity', color=color) 


# 添加 Operation Capability 0.05 的水平线
ax1.axhline(y=0.05, color='tab:blue', linestyle='--', alpha=0.5, linewidth=1,label='Operation Capability=0.05')
# 添加 Connectivity 0.05 的水平线
ax2.axhline(y=0.05, color='tab:red', linestyle='--', alpha=0.5, linewidth=1,label='Connectivity=0.05')

# 找到第一个Connectivity <= 0.05的点
zero_index_c = np.where(np.array(PMA_C) <= 0.05)[0]
if len(zero_index_c) > 0:
    first_zero_index_c = zero_index_c[0]
    x_zero_c = x[first_zero_index_c]
    y_zero_c = PMA_C[first_zero_index_c]
    
    # 在零点添加标记
    ax2.plot(x_zero_c, y_zero_c, 'ko', markersize=8, fillstyle='none', markeredgewidth=2)
    
    # 添加注释
    ax2.annotate(f'$f_N$={x_zero_c:.3f}', 
                 xy=(x_zero_c, y_zero_c),
                 xytext=(x_zero_c + 0.05, y_zero_c + 0.1),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=9)
    
    # 添加垂直线
    ax1.axvline(x=x_zero_c, color='gray', linestyle=':', alpha=0.7)

# 找到第一个Operation Capability <= 0.05的点
zero_index_p = np.where(np.array(PMA) <= 0.05)[0]
if len(zero_index_p) > 0:
    first_zero_index_p = zero_index_p[0]
    x_zero_p = x[first_zero_index_p]
    y_zero_p = PMA[first_zero_index_p]
    
    # 在零点添加标记
    ax1.plot(x_zero_p, y_zero_p, 'ks', markersize=8, fillstyle='none', markeredgewidth=2)
    
    # 添加注释
    ax1.annotate(f'$f_N$={x_zero_p:.3f}', 
                 xy=(x_zero_p, y_zero_p),
                 xytext=(x_zero_p + 0.05, y_zero_p + 0.1),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=9)
    
    # 添加垂直线
    ax1.axvline(x=x_zero_p, color='gray', linestyle=':', alpha=0.7)
               
# 添加标题和图例
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')


ax1.set_xlabel(r'$f_{N}$')  # 设置横坐标标签
# 保存图形并指定 dpi 参数
plt.savefig('pk-PMA', dpi=300)
# 显示图形
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_palette("deep")

# 数据
categories = ['-', 'RNC', 'MSSNC', 'MFSNC']

# 每组数据
without = [PMA_ANOC, RA_ANOC, PSA_ANOC, PDA_ANOC, PIA_ANOC]
RNC = [PMA_RNC_ANOC, RA_RNC_ANOC, PSA_RNC_ANOC, PDA_RNC_ANOC, PIA_RNC_ANOC]
MSNC = [PMA_MSNC_ANOC, RA_MSNC_ANOC, PSA_MSNC_ANOC, PDA_MSNC_ANOC, PIA_MSNC_ANOC]
MCNC = [PMA_MCNC_ANOC, RA_MCNC_ANOC, PSA_MCNC_ANOC, PDA_MCNC_ANOC, PIA_MCNC_ANOC]

# 组内标签
labels = ['PMA', 'RA', 'PSA', 'PDA', 'PIA']

# 组数
num_groups = len(categories)

# 绘图设置
fig, ax = plt.subplots(figsize=(8, 6))  # 设置尺寸为 (8, 6)
index = np.arange(num_groups)
bar_width = 0.15
opacity = 0.8

# 绘制每个柱状图并添加数值标签
for i in range(len(labels)):
    bars = plt.bar(index + i * bar_width, [without[i], RNC[i], MSNC[i], MCNC[i]], bar_width,
                   alpha=opacity,
                   label=labels[i],
                   edgecolor='black')  # 添加黑色外边框
    
    # 为每个柱子添加数值标签
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom', fontsize=8)

# 添加标签和标题
plt.xlabel('Collaborative reconfiguration strategy')
plt.ylabel('ANOC')
plt.xticks(index + bar_width, categories)
plt.legend()

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.3, which='both')

# 显示图形
plt.tight_layout()
# 保存图形并指定 dpi 参数
plt.savefig('bar-Cooperative Reconfiguration Strategy', dpi=300, bbox_inches='tight')
plt.show()
```
