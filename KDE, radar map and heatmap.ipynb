```python
import pandas as pd

# 读取Excel文件
excel_file = 'results2.xlsx'

# 创建一个空列表，用于存储每个子表格的平均值
averages = []

# 读取Excel文件中的所有子表格
xls = pd.ExcelFile(excel_file)

# 遍历所有sheet
for sheet_name in xls.sheet_names:
    # 读取每个子表格
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    
    # 提取第二行数据（Python的索引是从0开始的，所以第二行的索引是1）
    second_row = df.iloc[0]  # 获取第二行数据
    
    # 计算第二行的平均值
    average_value = second_row.mean()  # 计算平均值
    
    # 将平均值和子表格名称一起存入averages列表
    averages.append({'Sheet': sheet_name, 'Average': average_value})

# 将平均值数据转换为DataFrame
averages_df = pd.DataFrame(averages)

# 将计算的平均值结果保存到新的Excel文件
averages_df.to_excel('average2.xlsx', index=False)

print("计算完成，平均值已导出到average2.xlsx")
```

```python
#修改后
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
# 设置全局字体为'Times New Roman'
matplotlib.rcParams['font.family'] = 'Times New Roman'

# 读取 Excel 文件
file_path = 'results2.xlsx'
xls = pd.ExcelFile(file_path)

# # 设置绘图风格 - 使用更现代的Seaborn主题
# sns.set_style("whitegrid")


# 确定需要绘制的工作表对
num_sheets = len(xls.sheet_names)

# 每对工作表绘制一张图
for i in range(20):  # 只绘制到第20对工作表
    fig, ax = plt.subplots(figsize=(6, 4))  # 增加图形尺寸
    
    # 读取数据
    df1 = pd.read_excel(xls, sheet_name=i)
    values1 = df1.iloc[0].values
    
    if (i + 20) < num_sheets:
        df2 = pd.read_excel(xls, sheet_name=i + 20)
        values2 = df2.iloc[0].values
    else:
        print(f'工作表 {i + 20} 不存在。')
        continue
        
    # 获取工作表名称
    sheet_name1 = xls.sheet_names[i]
    sheet_name2 = xls.sheet_names[i + 20]
    
    # ===================== 统计验证 =====================
    # 1. 描述性统计
    desc1 = (f"{sheet_name1}\n"
             f"Mean: {np.mean(values1):.4f}\n"
             f"SD: {np.std(values1):.4f}\n"
             f"95% CI: [{np.mean(values1)-1.96*np.std(values1)/np.sqrt(len(values1)):.4f}, "
             f"{np.mean(values1)+1.96*np.std(values1)/np.sqrt(len(values1)):.4f}]")
    
    desc2 = (f"{sheet_name2}\n"
             f"Mean: {np.mean(values2):.4f}\n"
             f"SD: {np.std(values2):.4f}\n"
             f"95% CI: [{np.mean(values2)-1.96*np.std(values2)/np.sqrt(len(values2)):.4f}, "
             f"{np.mean(values2)+1.96*np.std(values2)/np.sqrt(len(values2)):.4f}]")
    
    # 2. 统计检验
    stats_text = ""
    
    # 方差分析 (ANOVA)
    try:
        _, anova_p = stats.f_oneway(values1, values2)
        anova_sig = "***" if anova_p < 0.001 else "**" if anova_p < 0.01 else "*" if anova_p < 0.05 else "ns"
        stats_text += f"ANOVA: p = {anova_p:.4f} {anova_sig}\n"
    except Exception as e:
        stats_text += f"ANOVA failed\n"
    
    # Kolmogorov-Smirnov 检验
    try:
        _, ks_p = stats.ks_2samp(values1, values2)
        ks_sig = "***" if ks_p < 0.001 else "**" if ks_p < 0.01 else "*" if ks_p < 0.05 else "ns"
        stats_text += f"KS-test: p = {ks_p:.4f} {ks_sig}"
    except Exception as e:
        stats_text += f"KS-test failed"
    
    # ===================== 绘图部分 =====================

    # 绘制概率密度曲线
    if len(values1) >= 3:
        sns.kdeplot(values1, label=sheet_name1, color='#1f77b4', shade=True)
    else:
        print(f'工作表{sheet_name1} 中的数值数量不正确:', len(values1))

    if len(values2) >= 3:
        sns.kdeplot(values2, label=sheet_name2, color='#d62829', shade=True)       
    else:
        print(f'工作表 {sheet_name2} 中的数值数量不正确:', len(values2))

    # 设置坐标轴范围
    plt.xlim(0, 0.75)
    plt.ylim(0, 280)

    
    # 设置标签和标题
  # 设置图例和标签
    plt.ylabel('Density')
    plt.xlabel('ANC or ANOC')
    
    # 添加统计信息框 - 使用更美观的布局
    # 第一组数据统计框
    plt.annotate(desc1, xy=(0.65, 0.65), xycoords='axes fraction', 
                bbox=dict(boxstyle="round,pad=0.4", fc="#d6eaf8", ec="#3498db", alpha=0.9),
                fontsize=10, fontweight='bold', color='#2c3e50')
    
    # 第二组数据统计框
    plt.annotate(desc2, xy=(0.65, 0.40), xycoords='axes fraction', 
                bbox=dict(boxstyle="round,pad=0.4", fc="#fadbd8", ec="#e74c3c", alpha=0.9),
                fontsize=10, fontweight='bold', color='#2c3e50')
    
    # 统计检验结果框
    plt.annotate(stats_text, xy=(0.65, 0.23), xycoords='axes fraction', 
                bbox=dict(boxstyle="round,pad=0.4", fc="#f9e79f", ec="#f1c40f", alpha=0.9),
                fontsize=10, fontweight='bold', color='#2c3e50')
    
    
    # 添加网格线增强可读性
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # 调整布局
    plt.tight_layout(rect=[0, 0, 0.95, 1])  # 为右侧信息框留出空间
    
    # 保存图像
    plt.savefig(f'kde_plot_{i + 1}.png', bbox_inches='tight', dpi=300)
    plt.close()  # 关闭当前图像
```
